### Question 1:
How many unique opponents have we played against? 
How many times have we played against each opponent?
What is the Illini win-loss record against Purdue and Chicago, (that is Win - Lost)? 
How many homecoming games have we played?
What is our Win-Loss record among homecoming games, (that is Win - Lost)?

```python


group = df.groupby('Opponent')
grouped_count = group.agg('count').reset_index()
uniq_opponent = len(grouped_count['Opponent'])

times_each = df['Opponent'].value_counts()


df_purdue = df[df['Opponent'] == 'Purdue']
df_chicago = df[df['Opponent'] == 'Chicago']
result_p = df_purdue['Result'].value_counts()
result_c = df_chicago['Result'].value_counts()

wins_lost_purdue = result_p[0] - result_p[1]
wins_lost_chicago = result_c[0] - result_c[1]

df_hc = df[df['Note'] == 'Homecoming']
num_homecoming = len(df_hc)

result_hc = df_hc['Result'].value_counts()
wins_lost_homecoming = result_hc[1] - result_hc[0]



times_each
```
