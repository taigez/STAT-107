## Dota 2 The International 2019 Hero performance analysis

The International 9, also known as TI9, has the biggest prize pool in all eSports tournaments, finalizing at US$34,300,000. The champion of this tournament takes away 15million together with the Agies of Immortality as their trophy. 

Let's analyze the various heroes used throughout this tournament to see what are the conditions that provided us with a higher probability to win 15million dollars in **Year 2020**. 

The following data is taken from **datdota.com** 

```python
import pandas as pd
df = pd.read_csv('hero_performance.csv')
df
```

## Heroes with the highest winrate 

Looking for the hero with the highest winrate, and picking them next year  is probably the most direct approach to increase our chances of winning.

However, some heroes are only picked once or twice throughout the whole competition with a 100% winrate. For such heroes, I believe they are not the deciding factor for their win. Therefore, I will exclude all the heroes with **less than 10 games** in total for this analysis.

```python
df_ten =  df[df['Total Count'] >= 10]
df_wr = df_ten.nlargest(5,'Winrate')
df_wr
```

## Heroes with most kills

We are also looking for heroes with the most game impacts. Kill death ratio(KDA) is something that take kills, deaths, and assists into consideration and it is a pretty good indicator of game impact. \

**KDA = (Kills + Assists) / Deaths**

Let's calculate KDA for all heros and find the heroes with the highest GPA.

```python
df['KDA'] = (df['Kills'] + df['Assists']) / df['Deaths']
df
```

```python
df_kda = df.nlargest(5,'KDA')
df_kda
```

## Relationship between Average Levels and Winrate

Does higher levels give us better winrate?

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model = model.fit(df[['LVL']], df['Winrate'])
model.coef_
```
